```python
#Create delta_distance data:

#delta_distance = np.zeros((len(distance_data),1))

delta_distance = list()

print(delta_distance)

changeInPreviousTrial = 0
for idx, distanceValue in enumerate(distance_data):

    delta_distance.append(abs(distanceValue - distance_data[0]))

print(delta_distance)

if len(delta_distance) != len(distance_data):
    raise Exception("delta_distance length does not match distance_data!!!!")
```

```python
force_values = list()

print(force_values)

for idx, massValue in enumerate(cal_mass):
    forceValueCal = round((massValue * 9.8), 4)

    #TODO: need that value to have 0.000 digits
    if forceValueCal == 0:
        pass
        #forceValueCal = float(format(str(forceValueCal), '.3f'))
    #bigger than ten because that's when it begins two digits before . and messes up sig digs.
    if forceValueCal >= 10:

        forceValueCal = round(forceValueCal, 2)
        print("changed value: {0}".format(forceValueCal))
    
    force_values.append(forceValueCal)




print(force_values)
if len(force_values) != len(cal_mass):
    raise Exception("force_values length does not match cal_mass!!!!")
```

```python
CompleteDataDict = {
    "Total Distance (mm)" : distance_data,
    "Change in Distance (mm)" : delta_distance,
    "Force (N)" : force_values
}

tableCompleteData = pd.DataFrame(CompleteDataDict)

#Step 1 complete.
tableCompleteData

```
